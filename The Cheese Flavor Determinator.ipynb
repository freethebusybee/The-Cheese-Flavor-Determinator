```python
import ipywidgets as widgets
from IPython.display import display, Markdown, HTML
import pandas as pd

# Load your dataset
df = pd.read_excel('/Users/farihasyed/Downloads/Cheese.xlsx')

# Clean up any leading or trailing whitespace in columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Define your top criteria
top_milk_types = ['cow', 'goat']
top_countries = ['United States', 'France', 'Italy', 'Netherlands']
top_flavors = ['creamy', 'nutty', 'sharp','tangy']

# Create dropdowns for user preferences
milk_type = widgets.Dropdown(options=top_milk_types, description='Milk Type:')
country = widgets.Dropdown(options=top_countries, description='Country:')
flavor = widgets.Dropdown(options=top_flavors, description='Flavor:')

# Display the widgets
display(milk_type, country, flavor)

# Output widget for displaying results
output = widgets.Output()
display(output)

def display_results(change=None):
    # Clear previous output
    with output:
        output.clear_output(wait=True)
        
        # Filter the DataFrame based on selected criteria
        filtered_cheeses = df[
            (df['milk'].str.lower() == milk_type.value.lower()) &
            (df['country'].str.lower() == country.value.lower()) &
            (df['flavor'].str.lower() == flavor.value.lower())
        ]
        
        # Display results
        if not filtered_cheeses.empty:
            # Prepare output
            display(HTML("<h3>Cheeses based on your preferences:</h3>"))
            for index, row in filtered_cheeses.iterrows():
                display(Markdown(f"- {row['cheese']}: [{row['url']}]({row['url']})"))
        else:
            print("No cheeses found that meet the criteria.")

# Attach the function to observe changes in dropdown values
milk_type.observe(display_results, names='value')
country.observe(display_results, names='value')
flavor.observe(display_results, names='value')

# Display initial results based on default selections
display_results()
```
